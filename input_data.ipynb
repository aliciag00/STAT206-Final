```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
countries = pd.read_csv("CountriesSD.csv")
summer = pd.read_csv("SummerSD.csv")
winter = pd.read_csv("WinterSD.csv")
```

```python
# for each country now lets split the data up between men and women and then fort in descending order to get the most recent games for each. 
summer_male_olympians = summer[summer['Gender'] == 'Men']
summer_female_olympians = summer[summer['Gender'] == 'Women']

summer_male_olympians.sort_values(by='Year', ascending=False).head(15)
```

```python
summer_female_olympians.sort_values(by='Year', ascending=False).head(15)
```

```python
#no I want to plot the relationship betweeen the gender and number of medals won by women vs men

#this code below will count it for the summer olympics and then plot it as a bar graph.

gender_counts = summer['Gender'].value_counts()
gender_counts.plot(kind='bar')
plt.title('Number of Medals Won by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Medals')      
```

```python
#now to split it up into the winter olympics and do the same thing

winter_male_olympians = winter[winter['Gender'] == 'Male']
winter_male_olympians.sort_values(by='Gender', ascending=False).head(15)

winter_male_olympians.head(15)
```
